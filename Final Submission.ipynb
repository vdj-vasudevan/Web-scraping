```python
import pandas as pd
import requests
import bs4
```

```python
res= requests.get('https://store.steampowered.com/games/')
res
soup = bs4.BeautifulSoup(res.text,'lxml')
Top5_game=[]
print("Top 5 Games which are New & Trending")
for i in soup.select('.tab_item_name',limit=5):
    Top5_game.append(i.text)
for j in Top5_game:
    print(f"🎮",j)
# Loop to Fetch the Urls of the Games from 'https://store.steampowered.com/games/' 
top_url=[]
for i in range(5):
    for div in soup.findAll('div', {'class': 'peeking_carousel'},limit=1):
        top_1 = div.findAll('a')[i]
#creating a list of urls
        top_url.append(top_1)
#creating a blank list of all the Columns  
Positive_review=[]
developer=[]
summary=[]
publisher=[]
msr=[]
rsr=[]
rev_each_pg=[]
#creating Loop to Generate top5 games url`
for url in top_url:
#creating Request for each call by the help of top_urls list    
    res= requests.get(url.attrs['href'])
# creating soup object
    soup = bs4.BeautifulSoup(res.text,'lxml')
    rev=[]
    list_per=[]
#creating list_per to Extract the Posivive Review details
    try:
        for dev in soup.find_all('span', {'class':'nonresponsive_hidden responsive_reviewdesc'}):
            list_per.append(dev.text.strip().replace('%','').replace(',',''))
        rev=list_per[-1]
        rev
    except:
        pass
#Use of isdigit() to Extract numbers from text
    try:
        ans=[int(i) for i in rev.split() if i.isdigit()]
        str(ans)
        p_review=round((ans[0]*ans[1])/100) 
        Positive_review.append(p_review)
    except:
        Positive_review.append(rev)
        pass
#creating developer to Extract the Developer details  
    dev_list=[]
    for devel in soup.find('div', {'id':'developers_list'}).find_all('a'):
        dev_list.append(devel.text)
    developer.append(dev_list)   
#creating publisher to Extract the Publisher details 
    summary=[]
    devp=soup.find_all('div', {'class':'dev_row'},limit=2)
    for j in devp:
        summary.append(str(j.text).strip().replace('\n',''))
    summary[1]=summary[1][summary[1].find(':')+1:]
    publisher.append(summary[1])
#Creating msr_i to extract list of MINIMUM SYSTEM REQUIREMENTS for each game
    msr_i=[]
#Creating rsr_i to extract list of RECOMMENDED SYSTEM REQUIREMENTS for each game
    rsr_i=[]
#Use of try:except:pass since all web page don't consist of the System details in same Class
    try:
        data1 = soup.find('div',attrs={'class':'game_area_sys_req_leftCol'}).find('ul',attrs={'class':'bb_ul'})
    except:
        pass
    try:
        data1 = soup.find('div',attrs={'class':'game_area_sys_req_full'}).find('ul',attrs={'class':'bb_ul'})
    except:
        pass
    msr_i.append("-->MINIMUM SYSTEM REQUIREMENTS-->:")
    for li in data1.find_all("li"):
        msr_i.append(li.text)

    try:
        data1 = soup.find('div',attrs={'class':'game_area_sys_req_rightCol'}).find('ul',attrs={'class':'bb_ul'})
        rsr_i.append("-->RECOMMENDED SYSTEM REQUIREMENTS-->:")
        for li in data1.find_all("li",limit=6):
            rsr_i.append(li.text)
    except:
        rsr_i.append("-->RECOMMENDED SYSTEM REQUIREMENTS-->: Nothing is Recommended")
        pass   
    
    msr.append(msr_i)
    rsr.append(rsr_i)
    
    Info  = (url.attrs['href']).split('?')[0]   
#Extracting Game ID from each url 
    Id = Info.split('/')[-3]
#Using the Static Url to Changing only ID to Extract Reviews:    
    review_url=(f'https://steamcommunity.com/app/'+Id+'/reviews/?browsefilter=toprated&snr=1_5_100010_&filterLanguage=all')
    
    res_1= requests.get(review_url)

    soup = bs4.BeautifulSoup(res_1.text,'lxml')
#creating rev_each_pg to Extract the Review of Each Game:    
    rev_pg=[]
    on_rev=[]
    for i in soup.find_all('div',attrs={'class':'apphub_CardTextContent'}):
        on_rev.append((i.text).strip())
    on_rev
    for i,j in enumerate(on_rev):
        temp=j 
        rev=[i+1],temp[temp.rfind('\t')+1:]
        rev_pg.append(rev)
    rev_each_pg.append(rev_pg)
        
#print(Positive_review)
#print(developer)
#print(publisher)
#print(msr)
#print(rsr)
#print(rev_each_pg)
```

```python
# Creating new List by appending Each Game Detail's list in it
Each_game=[]
# Creating the Index column First and appending it to the list first
index=['Game/ Description','Positive_review','Developer','Publisher','Min System Requirement','Recomended system Requirement', '10 Reviews']
Each_game.append(index)
for i in range(5):
    game=[]
    game.append(Top5_game[i])
    game.append(Positive_review[i])
    game.append(developer[i])
    game.append(publisher[i])
    game.append(msr[i])
    game.append(rsr[i])
    game.append(rev_each_pg[i])
    Each_game.append(game)
Each_game
```

```python
# Creating a DataFrame Using the Each_game list
df_game=pd.DataFrame(list(zip(Each_game[1],Each_game[2],Each_game[3],Each_game[4],Each_game[5])),
               columns =['Game 1', 'Game 2','Game 3','Game 4','Game 5'],index=Each_game[0])
```

```python
##only use Once
df_game.columns = df_game.iloc[0]
df_game = df_game[1:]
```
